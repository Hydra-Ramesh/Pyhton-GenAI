## Sets

Sets are unordered collections of unique elements. They are defined using curly braces {} or the set() constructor. Sets are mutable, meaning you can add or remove elements from them, but they do not allow duplicate values.

```python
# Creating a Set
my_set = {1, 2, 3, 4, 5}
print(my_set)
print(type(my_set))
```

```python
my_new_set = set()
print(my_new_set)
print(type(my_new_set))
```

```python
my_new_set={1, 2, 3, 4, 5}
print(my_new_set)
```

```python
my_list = [9, 8, 7, 6, 5]
my_new_set = set(my_list)
print(my_new_set)
```

```python
#Adding elements to a Set
my_new_set.add(10)
print(my_new_set)
```

```python
my_new_set.remove(6)
print(my_new_set)
```

```python
my_new_set.remove(6)
print(my_new_set)
```

```python
my_new_set.add(10)
```

```python
print(my_new_set)
```

```python
# Discarding duplicates
my_new_set.discard(10)
print(my_new_set)
```

```python
# Pop method
my_new_set.pop()
print(my_new_set)
```

```python
# Clearing a Set
my_new_set.clear()
print(my_new_set)
```

```python
num_set= {1, 2, 3, 4, 5}
print(num_set)
print(3 in num_set)
print(6 in num_set)
```

```python
set1= {1, 2, 3}
set2= {3, 4, 5}
u_set= set1.union(set2)
print(u_set)
```

```python
i_set= set1.intersection(set2)
print(i_set)
```

```python
set1.difference(set2)
```

```python
s1={1,2,3,4,5}
s2={3,4,5}
print(s1.issubset(s2))
print(s2.issubset(s1))
```

```python
ss1={1,2,3,4,5}
ss2={3,4,5}
print(ss1.issuperset(ss2))
print(ss2.issuperset(ss1))
```

```python
# Counting Unique Words in a text
text = "Hello world! Hello everyone. Welcome to the world of Python."
words = text.split()
unique_words = set(words)
print(unique_words)
```
