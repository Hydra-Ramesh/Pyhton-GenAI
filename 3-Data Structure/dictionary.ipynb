## Dictionary

Dictionary is a collection of key-value pairs. It is an unordered, mutable, and indexed data structure in Python. Each key is unique and maps to a value. Dictionaries are defined using curly braces `{}` with key-value pairs separated by a colon `:`.

user = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

```python
empty_dict = {}
print(empty_dict)
print(type(empty_dict))
```

```python
emp_dict = dict()
print(emp_dict)
print(type(emp_dict))
```

```python
user = {"name":"Anik","age":18,"city":"Midnapore"}
print(user)
```

```python
print(user["name"])
```

```python
print(user.get("age"))
```

```python
user["age"]=19
print(user)
```

```python
user["email"]="anik@example.com"
print(user)
```

```python
print(user.keys())
print(user.values())
```

```python
print(user.items())
```

```python
user1=user.copy()
```

```python
print(user1)
```

```python
user["name"]="Rishi"
print(user)
print(user1)
```

```python
user1["age"]=20
print(user)
print(user1)
```

```python
# Itrating through a dictionary
for key in user:
    print(key)
```

```python
for key in user:
    print(key,user[key])
```

```python
#Nested Dictionary
person={
    "name":"Anik",
    "age":18,
    "address":{
        "street":"123 Main St",
        "city":"Midnapore",
        "country":"India"
    }
}
print(person)
print(person["address"]["city"])
```

```python
d1={"a":1,"b":2}
d2={"c":3,"d":4}
merged_dict={**d1,**d2}
print(merged_dict)
```
