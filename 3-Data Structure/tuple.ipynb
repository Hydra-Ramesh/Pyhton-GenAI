## Tuple

The Tuple data structure is an ordered collection of elements that can hold a fixed number of items. It is similar to a list, but unlike lists, tuples are immutable, meaning that once they are created, their contents cannot be changed.

```python
# Creating Tuple
created_tuple=()
print(created_tuple)
print(type(created_tuple))
```

```python
# Conversion of List to Tuple
nums=[1,2,3,4,5]
converted_tuple=tuple(nums)
print(converted_tuple)
print(type(converted_tuple))
print(nums)
print(type(nums))
```

```python
mixed_tuple=(1,"Hello",3.14,True)
print(mixed_tuple)
```

```python
numbers_tuple=(1,2,3,4,5)
print(numbers_tuple[0])
print(numbers_tuple[1])
print(numbers_tuple[2])
print(numbers_tuple[3])
print(numbers_tuple[4])
```

```python
print(numbers_tuple[-1])
print(numbers_tuple[-2])
```

```python
print(numbers_tuple[0:3])
```

```python
print(numbers_tuple[::-1])
```

```python
t1=(1,2,3)
t2=(4,5,6)
t3=t1+t2
print(t3)
```

```python
t3*3
```

```python
members=("Alice","Bob","Charlie")
print(members[0])
members[0]="David"
```

```python
# Some built-in functions for tuple
names=("Anik","Saptarshi","Ramesh")
print(len(names))
```

```python
print(names.count("Anik"))
```

```python
print(names.index("Saptarshi"))
```

```python
# Packing and Unpacking of Tuple
packed_tuple=1,"Hello",3.14,True
print(packed_tuple)
```

```python
a,b,c,d=packed_tuple
print(a)
print(b)
print(c)
print(d)
```

```python
# Unpacking with * operator
num=(1,2,3,4,5,6,7,8,9)
first,*middle,last=num
print(first)
print(middle)
print(last)
```

```python
# Nested Tuple
nested_tuple=(1,2,(3,4),5)
print(nested_tuple)
```

```python
# List inside Tuple
list_tuple=(1,2,[3,4],5)
print(list_tuple)
```

```python
# Tupil inside List
list_with_tuple=[1,2,(3,4),5]
print(list_with_tuple)
```

```python
tuple2D=((1,2),(3,4),(5,6))
print(tuple2D)
```

```python
for row in tuple2D:
    for element in row:
        print(element)
    print()
```
