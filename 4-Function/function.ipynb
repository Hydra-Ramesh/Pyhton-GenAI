## Function

The function is a block of code which only runs when it is called. You can pass data, known as parameters, into a function. A function can return data as a result.

```python
# Syntex of Function
def functiona_name(parameters):
    """docstring"""
    # body of function
    return expression
```

```python
# Odd or Even
def odd_even(num):
    if num%2==0:
        return "Even"
    else:
        return "Odd"
```

```python
odd_even(5)
```

```python
odd_even(4)
```

```python
def addnum(a,b):
    return a+b
```

```python
addnum(2,3)
```

```python
def greetings(name):
    return "Hello "+name+" How are you?"
```

```python
greetings("Anik")
```

```python
greetings()
```

```python
#default parameter
def another_greeting(name="User"):
    return "Hello "+name+" How are you?"
```

```python
another_greeting("Saptarshi")
```

```python
another_greeting()
```

```python
# Variable length argument
def printing(a,b,c,d):
    print(a)
    print(b)
    print(c)
    print(d)
printing(1,2,3,4)
```

```python
def another_printing(*nums):
    for i in nums:
        print(i)

another_printing(1,2,3,4,5,6,7,8,9,"Anik","Saptarshi")
```

```python
# Keyword Argument
def keyword_printing(name,age):
    print("Name: "+name)
    print("Age: "+str(age))
keyword_printing("Anik",24)
```

```python
def another_keyword_printing(**data):
    for key,value in data.items():
        print(key+" : "+str(value))

another_keyword_printing(name="Anik", age=24, city="Kolkata", country="India")
```
